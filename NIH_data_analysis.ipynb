```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('RCDCFundingSummary_09072023.xlsx',skiprows=1)
df.drop(df.tail(48).index,inplace=True)
df.columns=df.iloc[0]
df=df[1:]
df.replace('+',np.nan,inplace=True)
df.replace('*',np.nan,inplace=True)
df.replace('-',0,inplace=True)
df
```

```python
df.infer_objects()
print(df.dtypes)
```

```python
df.nlargest(n=40,columns=['2023 Estimated'])
```

```python
df_top_30=df[(df['2020 US Mortality 19']!=0)|(df['2020 US Prevalence SE 19']!=0)].nlargest(n=30,columns=['2023 Estimated'])
df_top_30
```

```python
plt.barh(df_top_30[df.columns[0]][0:11].to_numpy(),df_top_30[2022][0:11].to_numpy())
plt.title('Top NIH fundings for different disease')
plt.xlabel('Millions USD')
plt.show()
```

```python
for i in range(10):
    plt.plot(df_top_30[df.columns[12:20]].iloc[i].to_numpy(),label=df_top_30[df.columns[0]].iloc[i],marker='o',
            linestyle='--')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(np.arange(8),df.columns[12:20],rotation=30)
plt.show()
```

```python
df_top_areas=df[(df['2020 US Mortality 19']==0)&(df['2020 US Prevalence SE 19']==0)&
                (~df[df.columns[0]].str.contains('Clinical','Research'))&(
                (df[df.columns[0]].str.contains('Diseases'))|
                    (df[df.columns[0]].str.contains('Disorder'))|
                    (df[df.columns[0]].str.contains('degenerative'))|
                    (df[df.columns[0]].str.contains('Disparities')))].nlargest(n=30,columns=[2022])
df_top_areas
```

```python
plt.barh(df_top_areas[df.columns[0]][0:11].to_numpy(),df_top_areas[2022][0:11].to_numpy())
plt.title('Top NIH fundings for different types of diseases')
plt.xlabel('Millions USD')
plt.show()
```

```python
for i in range(10):
    plt.plot(df_top_areas[df.columns[12:20]].iloc[i].to_numpy(),label=df_top_areas[df.columns[0]].iloc[i],marker='o',
            linestyle='--')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(np.arange(8),df.columns[12:20],rotation=30)
plt.show()
```
